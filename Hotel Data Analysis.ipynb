# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Loading the dataset

```python
df = pd.read_csv('hotel_bookings.csv')
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'],  dayfirst=True)
```

```python
for col in df.describe(include='object').columns :
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
df.drop(['agent', 'company'], axis=1, inplace= True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = df[df['adr'] < 5000]
```

# Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
plt.figure(figsize = (5,4))
plt.title('Reservation Status Count', fontsize = 20)
plt.bar(['Not Cancelled', 'Cancelled'], df['is_canceled'].value_counts(), width = 0.7)
plt.show
```

# Countplot - Hotel wise Reservation Status

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette = 'Blues')
plt.title('Reservation status in different hotels', size = 20)

legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['Not Cancelled','Cancelled'], bbox_to_anchor = (1,1)) 
plt.xlabel('hotel')
plt.ylabel('No. of reservations')
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

# Grouping based on Date and ADR

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()

plt.figure(figsize = (20, 8))
plt.title('Average Daily Rate in City & Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

# Monthwise reservation

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figure = (16,8))

ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data=df, palette = 'bright')

legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1)) 

plt.title('Reservation status per month', size = 20)
plt.xlabel('Month')
plt.ylabel('No. of reservations')
plt.legend(['Not Cancelled', 'Cancelled'])
plt.show()
```

# Checking ADR per month for cancelled bookings

```python
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 30)

# Group by month and reset the index
grouped_df = df[df['is_canceled'] == 1].groupby('month', as_index=False)[['adr']].sum()

# Plotting with explicit x and y arguments
sns.barplot(x='month', y='adr', data=grouped_df)

plt.show()
```

# Country-wise cancellation

```python
canceled_data = df[df['is_canceled'] == 1]
top_10_countries = canceled_data['country'].value_counts()[:10]

plt.figure(figsize = (8,8))
plt.title('Top 10 countries in cancellation', fontsize=20)
plt.pie(top_10_countries, autopct = '%.2f', labels = top_10_countries.index)
plt.show()
```

# Top market segment for cancelled bookings

```python
canceled_data['market_segment'].value_counts(normalize = True)
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date', inplace=True)

not_canceled_data = df[df['is_canceled'] ==0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace=True)
```

```python
plt.figure(figsize = (20,6))
plt.title('ADR')
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label='Not canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='Canceled')
plt.legend(fontsize = 20)

#This plot shows inconsistency
```

```python
# Making the above plot consistent
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date'] > '2016') & (canceled_df_adr['reservation_status_date'] < '2017-09')]
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] > '2016') & (not_canceled_df_adr['reservation_status_date'] < '2017-09')]

plt.figure(figsize = (20,6))
plt.title('ADR')
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label='Not canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label='Canceled')
plt.legend(fontsize = 20)
```
